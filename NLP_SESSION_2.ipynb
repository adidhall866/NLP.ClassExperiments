NLTK PIPELINE

```python
texts = ["""Wolverines, also known as skunk bears or nasty cats, can confront and kill animals as large as moose and bears. Their oily fur is resistant to frost, and they have a special rotated molar that allows them to easily rip apart meat that's been frozen solid. """]
for text in texts:
    sentences = nltk.sent_tokenize(text)
    for sentence in sentences:
        words=nltk.word_tokenize(sentence)
        tagged_words = nltk.pos_tag(words)
        print(tagged_words)
```
